```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
import os
plt.rcParams["figure.figsize"] = [16,8]
```

```
A = imread("/content/dog_image.jfif")
x = np.mean(A , -1) #grayscaling 

img = plt.imshow(x)
img.set_cmap("gray")
plt.axis("off")
plt.show()

```

```
U , S  , VT = np.linalg.svd(x , full_matrices= False) #economical SVD
S = np.diag(S)
```

```
VT.shape
```

```
j = 0
for r in (5 , 20 , 100 , 1000):  #rank
  Xapprox = U[: , :r] @ S[:r , :r] @ VT[:r , :]
  plt.figure(j+1)
  img = plt.imshow(Xapprox)
  img.set_cmap("gray")
  plt.axis("off")
  plt.title("rank = " + str(r))
  plt.show()

```

```
plt.figure(1)
plt.semilogy(S)
plt.title("singular values")
plt.show()

plt.figure(2)
plt.plot(np.cumsum(S / np.sum(S)))
plt.title("sigular values : cumulative sum")
plt.show()
```
